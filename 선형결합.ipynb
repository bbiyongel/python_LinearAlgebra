# 선형 결합

- 선형결합 : n개의 벡터 x1,x2,x3....,xn 와 n개의 스칼라 c1,c2,c3...,cn 가 있을 때 다음 식을 __'선형 결합(Linear Combination)'__ 이라고 한다.  
        c1x1 + c2x2 + c3x3 + .... + cnxn  

    벡터에 스칼라를 곱한뒤, 모두 더한것 .    
    
      
- 선형종속 : 스칼라 c1,c2,c3...,cn가 모두 0이 아닐 때, 아래 식을 만족하면 벡터 x1,x2,x3....,xn를 __'선형 종속(Linear dependent)'__ 라고 한다. 적어도 하나의 벡터를 다른 벡터의 조합으로 나타낼 수 있다는 의미이다.
        c1x1 + c2x2 + c3x3 + .... + cnxn = 0  
  
  
- 선형독립 : 반대로 개수 c1,c2,c3...,cn가 모두 0일 때에만 위의 식이 만족할 때 벡터 x1,x2,x3....,xn를 __'선형 독립(Linear independent)'__ 라고 한다. 하나의 벡터를 다른 벡터의 조합으로 나타낼 수 없다는 의미입니다. 
        c1x1 + c2x2 + c3x3 + .... + cnxn = 0   * (c1,c2,c3...,cn = 0)

### 선형종속

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = np.array([1,5,0])
b = np.array([4,4,0])
c = np.array([4,7,0])

fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection = '3d')

ax.quiver(0, 0, 0, a[0], a[1], a[2], color = 'black', arrow_length_ratio = 0.1)
ax.text(a[0], a[1], a[2], 'a', size = 15)

ax.quiver(0, 0, 0, b[0], b[1], b[2], color = 'black', arrow_length_ratio = 0.1)
ax.text(b[0], b[1], b[2], 'b', size = 15)

ax.quiver(0, 0, 0, c[0], c[1], c[2], color = 'blue', arrow_length_ratio = 0.1)
ax.text(c[0], c[1], c[2], 'c', size = 15)

ax.set_xlim(0,8)
ax.set_ylim(0,8)
ax.set_zlim(0,8)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=20, azim=5)
ax.grid()

#### 7. 보조선(눈금선) 나타내기
ax.set_axisbelow(True)

plt.show()
```

```python
# 벡터 c는 a, b의 선형 결합으로 나타낼 수 있기 때문에 선형 종속이다.
# (0.75)a + (0.81)b = c
# 벡터 c를 왼쪽으로 옮겨 보면 0이 아닌 개수를 가지는 벡터의 선형 결합이 됩니다. 
# (0.75)a + (0.81)b - (1.0)c = 0


import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = np.array([1,5,0])
b = np.array([4,4,0])
c = np.array([4,7,0])

a = a*0.75
b = b*0.81

fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')

ax.quiver(0, 0, 0, a[0], a[1], a[2], color = 'black', arrow_length_ratio = 0.1)
ax.text(a[0], a[1], a[2], 'a', size = 15)

ax.quiver(a[0], a[1], a[2], b[0], b[1], b[2], color = 'black', arrow_length_ratio = 0.1)
ax.text(b[0], b[1], b[2], 'b', size = 15)

ax.quiver(0, 0, 0, c[0], c[1], c[2], color = 'blue', arrow_length_ratio = 0.1)
ax.text(c[0], c[1], c[2], 'c', size = 15)

ax.set_xlim(0,8)
ax.set_ylim(0,8)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# ax.view_init (각도 돌려보기)
ax.view_init(45,45)
ax.grid()
ax.set_axisbelow(True)

plt.show()
```

### 선형 독립

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = np.array([10, 2, 0])
b = np.array([4, 5, 0])
c = np.array([0, 0, 5])

fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')

ax.quiver(0, 0, 0, a[0], a[1], a[2], color = 'black', arrow_length_ratio=0.1)
ax.text(a[0], a[1], a[2], 'a',)

ax.quiver(0, 0, 0, b[0], b[1], b[2], color = 'black', arrow_length_ratio=0.1)
ax.text(b[0], b[1], b[2], 'b',)

ax.quiver(0, 0, 0, c[0], c[1], c[2], color = 'blue', arrow_length_ratio=0.1)
ax.text(c[0], c[1], c[2], 'c',)

ax.set_xlim(0,11)
ax.set_ylim(0,11)
ax.set_zlim(0,11)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=20, azim=5)
ax.grid()
ax.set_axisbelow(True)

plt.show()
```

### 랭크

행렬을 구성하는 열벡터 중 선형 독립인 열벡터의 개수를 __'열랭크(column rank)'__ 라고 부르고 행렬을 구성하는 행벡터 중 선형 독립인 행벡터의 개수를 __'행랭크(row rank)'__ 라고 한다. 행렬에 대한 열랭크와 행랭크는 항상 같으므로 구별없이 __'rank'__ 라고 한다.

```python
# numpy에서는 matrix_rank함수를 통해 rank를 구할 수 있다.

A = np.array([[0,1,2],[1,2,1],[2,7,8]])
np.linalg.matrix_rank(A)
```
